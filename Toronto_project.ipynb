## First point

### Libraries 

```python
import pandas as pd
import numpy as np
```

```python
# Import dataframe
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
```

```python
df
```

## Cleaning the data

```python
[df[j][i][:3] for i in range(0,len(df),1) for j in range(0,len(df.columns),1)]
```

```python
b=pd.DataFrame([df[j][i][3:].split('(') for i in range(0,len(df),1) for j in range(0,len(df.columns),1)],columns=['borough','neighborhood','dont know'])
```

```python
a=pd.DataFrame([df[j][i][:3] for i in range(0,len(df),1) for j in range(0,len(df.columns),1)],columns=['post_code'])
```

```python
df_final=pd.concat([a,b],axis=1)
```

```python
df_final=df_final[df_final['borough']!='Not assigned'].reset_index(drop=True)
```

```python
df_final['neighborhood']=df_final['neighborhood'].apply(lambda x : x if x == None else x.replace(')',''))
```

```python
df_final[df_final['neighborhood'].isnull()]=df_final['borough']
```

```python
df_final=df_final[df_final['post_code']!="Queen's Park / Ontario Provincial Government"]
```

## Second part

```python
lat_long_postal_code=pd.read_csv('http://cocl.us/Geospatial_data')
```

```python
lat_long_postal_code.head()
```

```python
df_final=df_final.merge(lat_long_postal_code,left_on='post_code',right_on='Postal Code',how='left').drop(columns='Postal Code')
```

```python
df_final.head()
```

```python
df_final.head()
```

```python
df=df_final[df_final['borough'].str.contains('Toronto')].copy()
```

```python
df.head()
```

## Third part

```python
#map library
import folium
```

```python
neighborhood=df['neighborhood']
borough=df['borough']
```

```python
#Import geopy for Toronto Latitude and longitude
from geopy.geocoders import Nominatim
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['borough'], df['neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

# KMeans model

```python
# ML libraries
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
```

## See the best K for KMeans

```python
model=KMeans()
```

```python
visualizer = KElbowVisualizer(model,k=(1,12))
```

```python
X=df[['Latitude','Longitude']]
```

```python
visualizer.fit(X)
visualizer.show()
```

```python
# total number of cluster
kclusters = 4
# KMeans model
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(X)
#labels
kmeans.labels_
```

```python
df['Cluster Labels']=kmeans.labels_
```

```python
#Visualization libraries
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df['Latitude'], df['Longitude'], df['neighborhood'], df['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
