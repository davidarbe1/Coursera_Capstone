## First point

### Libraries 

```python
import pandas as pd
import numpy as np
```

```python
# Import dataframe
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
```

```python
df
```

## Cleaning the data

```python
[df[j][i][:3] for i in range(0,len(df),1) for j in range(0,len(df.columns),1)]
```

```python
b=pd.DataFrame([df[j][i][3:].split('(') for i in range(0,len(df),1) for j in range(0,len(df.columns),1)],columns=['borough','neighborhood','dont know'])
```

```python
a=pd.DataFrame([df[j][i][:3] for i in range(0,len(df),1) for j in range(0,len(df.columns),1)],columns=['post_code'])
```

```python
df_final=pd.concat([a,b],axis=1)
```

```python
df_final=df_final[df_final['borough']!='Not assigned'].reset_index(drop=True)
```

```python
df_final['neighborhood']=df_final['neighborhood'].apply(lambda x : x if x == None else x.replace(')',''))
```

```python
df_final[df_final['neighborhood'].isnull()]=df_final['borough']
```

```python
df_final=df_final[df_final['post_code']!="Queen's Park / Ontario Provincial Government"]
```

## Second part

```python
lat_long_postal_code=pd.read_csv('http://cocl.us/Geospatial_data')
```

```python
lat_long_postal_code.head()
```

```python
df_final=df_final.merge(lat_long_postal_code,left_on='post_code',right_on='Postal Code',how='left').drop(columns='Postal Code')
```
